```python
!pip install google-generativeai
!pip install python-dotenv
import os
import google.generativeai as genai
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv("key.env")

# Access API key
api_key = os.getenv("GEMINI_API_KEY")

# Ensure API key is loaded
if not api_key:
    raise ValueError("API key not found! Make sure .env file is correctly set up.")

# Configure GenAI
genai.configure(api_key=api_key)
```

```python

def upload_to_gemini(path, mime_type=None):
  """Uploads the given file to Gemini.

  See https://ai.google.dev/gemini-api/docs/prompting_with_media
  """
  file = genai.upload_file(path, mime_type=mime_type)
  print(f"Uploaded file '{file.display_name}' as: {file.uri}")
  return file

```

```python

generation_config = {
  "temperature": 0.2,
  "top_p": 0.95,
  "top_k": 40,
  "max_output_tokens": 8192,
  "response_mime_type": "text/plain",
}
```

```python
model = genai.GenerativeModel(
  model_name="gemini-2.0-flash",
  generation_config=generation_config,
)
```

```python
from pathlib import Path

file_path_pdf = r"Prenuptial-Agreement.pdf"

report_name = Path(file_path_pdf).stem

files = [
  upload_to_gemini(file_path_pdf, mime_type="application/pdf"),
]
```

```python
prompt1="""
You are a multi-skilled legal assistant and data extraction expert. Using the uploaded PDF document as your only input, please perform the following tasks:

1. **Document Summary:**  
   - Summarize the document in one concise paragraph.
   - Focus on identifying the key parties involved, primary legal claims, relevant statutes or regulations, and any important dates or deadlines.

2. **Discrepancy Detection:**  
   - Analyze the document for any issues such as missing standard clauses, incorrect statutory references, or contradictory statements.
   - Provide a bullet-point list of discrepancies along with suggested corrections for each.

3. **Draft Generation:**  
   - Based on the content of the document, generate a draft legal document (for example, a complaint or motion).
   - Ensure the draft includes standard legal headings (e.g., Introduction, Parties, Jurisdiction, Claims, Prayer for Relief) and is formatted clearly and concisely.


Please output your final answer as plain text with the following clearly labeled sections:
- Summary:
- Discrepancies:
- Draft:

"""
```

```python
chat_session = model.start_chat(
  history=[
    {
      "role": "user",
      "parts": [
        files[0],
       prompt1
      ],
    },
    
   
  ],
  
)
```

```python
response = chat_session.send_message("can u list all the discreapnices and error in legalitities involved in here")
```
